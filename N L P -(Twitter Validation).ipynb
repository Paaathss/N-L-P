```python
import numpy as np
import pandas as np
df_col = ["ID","LOCATION","TARGET","TEXT"]
df =pd.read_csv('/content/drive/MyDrive/MACHINE LEARNING /NLP(natural language preprocessing)/twitter_validation.csv',names = df_col)
df.head()
```

```python
df.isna().sum()
```

```python
df["TARGET"].value_counts()
```

```python
df.drop(df.index[(df["TARGET"]=="Irrelevant")],axis = 0,inplace=True)
df["TARGET"].value_counts()
```

```python
df.head()
```

```python
df.reset_index(drop=True,inplace=True)
df     #drop the old index and assigning new index to that row
```

```python
df.drop(["ID","LOCATION"],axis = 1,inplace=True)
df
```

```python
df["TARGET"] = df["TARGET"].map({"Positive":1,"Negative":1,"Neutral":0})
df
```

```python
tweets = df.TEXT
tweets
```

```python
tweets = tweets.str.replace(r"[^a-zA-Z0-9]", " ")
tweets
```

```python
tweets = tweets.str.replace(r"[^a-zA-Z0-9]", " ",regex = True)
tweets
```

```python
from nltk.stem import SnowballStemmer
from nltk import TweetTokenizer
stemmer = SnowballStemmer("english")
tk = TweetTokenizer()
```

```python
tweets = tweets.apply(lambda x:[stemmer.stem(i.lower())for i in tk.tokenize(x)]).apply(lambda y:" ".join(y))
tweets
```

```python
from nltk.corpus import stopwords
import nltk
nltk.download("stopwords")

sw = set(stopwords.words("english"))
```

```python
tweets = tweets.apply(lambda x:[i for i in tk.tokenize(x) if i not in sw]).apply(lambda x:" ".join(x))
tweets
```

```python
tweets.shape
```

```python
tweets
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
np.vectorizer = TfidfVectorizer()
train_vec = np.vectorizer.fit_transform(tweets)
print(train_vec)
```

```python
train_vec.shape
```

```python
x = train_vec
y = df["TARGET"].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.30,random_state=0)
```

**MODEL  BUILDING**

```python
from sklearn.svm import SVC
from sklearn.naive_bayes import MultinomialNB
from sklearn.ensemble import RandomForestClassifier,AdaBoostClassifier
svm_model = SVC()
nm_model = MultinomialNB()
rf_model = RandomForestClassifier()
ab_model = AdaBoostClassifier()

lst_model = [svm_model,nm_model,rf_model,ab_model]
```

```python
from sklearn.metrics import classification_report,ConfusionMatrixDisplay
```

```python
print(y_test.shape)
print(y_train.shape)
```

```python
from tkinter.constants import Y
for i in lst_model:
  print(i)
  i.fit(x_train,y_train)
  Y_pred = i.predict(x_test)
  print("*"*100)
  print(classification_report(y_test,Y_pred))
  print("*"*100)
  print(ConfusionMatrixDisplay.from_predictions(y_test,Y_pred))
```
